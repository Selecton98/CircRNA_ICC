```r
my_comparisons<-list(c("type1","type2"),c("type1","type3"),c("type2","type3"))
```

```r
options(repr.plot.height=5,repr.plot.width=3)
```

```r
ciber2
```

```r
library(ggpubr)
library(RColorBrewer)
p<-ggboxplot(ciber2, x="group", y="CD8_naive", fill = "group",order=c("type1","type2","type3"),palette =c("firebrick",'steelblue'))+
 stat_compare_means(comparisons = my_comparisons, label = "p.signif")+#label这里表示选择显著性标记（星号） 
 scale_fill_brewer(palette='Spectral')+
    stat_compare_means(label.y = 0)
p
```
