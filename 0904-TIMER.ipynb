```r
colnames(xcell2TIMER)
```

```r
library(ggpubr)
library(RColorBrewer)
p<-ggboxplot(xcell2TIMER, x="group", y="T cell CD8+_TIMER", fill = "group",order=c("type1","type2","type3"))+
 stat_compare_means(comparisons = my_comparisons, label = "p.signif")+#label这里表示选择显著性标记（星号） 
 scale_fill_brewer(palette='Spectral')  + 
    stat_compare_means(label.y = 0)
p
ggsave('CD8TIMER.pdf',height=5,width=3)
```
